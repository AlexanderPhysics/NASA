```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
import numpy as np
```

###Making a simple map

```python
# make sure the value of resolution is a lowercase L,
#  for 'low', not a numeral 1
my_map = Basemap(projection='ortho', # ortho -> Orthographic 
                 lat_0=50,           # lon_0 -> center of desired map domain (in degrees)
                 lon_0=-100,         # lat_0 -> center of desired map domain (in degrees)
                 resolution='l',     # l -> low resolution, h -> high resolution 
                 area_thresh=1000.0)
 
my_map.drawcoastlines() # why is basemap the Earth? Explore after tutorial 
 
plt.show()
```

##Adding detail

###Add Coast lines

```python
my_map.drawcoastlines()
my_map.drawcountries()
my_map.fillcontinents(color='coral')
 
plt.show()
```

###Draw globe boundary 

```python
my_map.drawcoastlines()
my_map.drawcountries()
my_map.fillcontinents(color='coral')
my_map.drawmapboundary()
 
plt.show()
```

###Draw lat and long lines


The np.arange() arguments tell where your latitude and longitude lines should begin and end, and how far apart they should be spaced.

###Ex 1

```python
my_map.drawcoastlines()
my_map.drawcountries()
my_map.fillcontinents(color='coral')
my_map.drawmapboundary()
 
my_map.drawmeridians(np.arange(0, 360, 30))
my_map.drawparallels(np.arange(-90, 90, 30))
 
plt.show()
```

###Ex 2

```python
my_map.drawcoastlines()
my_map.drawcountries()
my_map.fillcontinents(color='coral')
my_map.drawmapboundary()
 
my_map.drawmeridians(np.arange(0, 360, 10))
my_map.drawparallels(np.arange(-90, 90, 10))
 
plt.show()
```

###Adjust Globe Perspective

```python
my_map = Basemap(projection='ortho', 
                 lat_0=0, 
                 lon_0=-100,
                 resolution='l', 
                 area_thresh=1000.0)
 
my_map.drawcoastlines()
my_map.drawcountries()
my_map.fillcontinents(color='coral')
my_map.drawmapboundary()
 
my_map.drawmeridians(np.arange(0, 360, 30))
my_map.drawparallels(np.arange(-90, 90, 30))
 
plt.show()
```

###Change Type of Projection

 Change the projection type to robin. You should end up with a Robinson projection instead of a globe:

```python
# make sure the value of resolution is a lowercase L,
#  for 'low', not a numeral 1
my_map = Basemap(projection='robin', # robin -> Robinsion Projection
                 lat_0=0, 
                 lon_0=-100,
                 resolution='l', 
                 area_thresh=1000.0)
 
my_map.drawcoastlines()
my_map.drawcountries()
my_map.fillcontinents(color='coral')
my_map.drawmapboundary()
 
my_map.drawmeridians(np.arange(0, 360, 30))
my_map.drawparallels(np.arange(-90, 90, 30))
 
plt.show()
```

###Zooming In

```python
my_map = Basemap(projection='merc', # mercator projection
                 lat_0=57, 
                 lon_0=-135,
                 resolution = 'l', 
                 area_thresh = 1000.0,
                 llcrnrlon=-136.25, # lower left corner at 136.25 degrees west and 56 degrees north  
                 llcrnrlat=56,
                 urcrnrlon=-134.25, # upper right corner ...
                 urcrnrlat=57.75)
 
my_map.drawcoastlines()
my_map.drawcountries()
my_map.fillcontinents(color='coral')
my_map.drawmapboundary()
 
my_map.drawmeridians(np.arange(0, 360, 30))
my_map.drawparallels(np.arange(-90, 90, 30))
plt.title("Baranof Island in southeast Alaska")
plt.show()
```

###Change resolution to 'high'

```python
my_map = Basemap(projection='merc', 
                 lat_0=57, 
                 lon_0=-135,resolution = 'h', 
                 area_thresh = 1000.0,
                 llcrnrlon=-136.25, 
                 llcrnrlat=56,
                 urcrnrlon=-134.25, 
                 urcrnrlat=57.75)
 
my_map.drawcoastlines()
my_map.drawcountries()
my_map.fillcontinents(color='coral')
my_map.drawmapboundary()
plt.title("Baranof Island in southeast Alaska")
my_map.drawmeridians(np.arange(0, 360, 30))
my_map.drawparallels(np.arange(-90, 90, 30))
 
plt.show()
```


This is much better, but we're still missing an entire island to the west. This is because of the area_thresh setting. This setting specifies how large a feature must be in order to appear on the map. The current setting will only show features larger than 1000 square kilometers. This is a reasonable setting for a low-resolution map of the world, but it's a really bad choice for a small-scale map


Lets change that setting to 0.1, and see how much detail we get

```python
my_map = Basemap(projection='merc', 
                 lat_0=57, 
                 lon_0=-135,resolution = 'h', 
                 area_thresh = 0.1,
                 llcrnrlon=-136.25, 
                 llcrnrlat=56,
                 urcrnrlon=-134.25, 
                 urcrnrlat=57.75)
 
my_map.drawcoastlines()
my_map.drawcountries()
my_map.fillcontinents(color='coral')
my_map.drawmapboundary()
 
my_map.drawmeridians(np.arange(0, 360, 30))
my_map.drawparallels(np.arange(-90, 90, 30))

plt.title("Baranof Island in southeast Alaska")

plt.show()
```
